### Comparación entre Películas y Series en Netflix

En este gráfico mostramos la distribución del contenido en la plataforma.  
La mayoría corresponde a **películas**, mientras que las **series** representan una proporción menor.  
Esto refleja que Netflix, aunque es conocido por sus series originales, todavía tiene un catálogo más amplio de películas.


```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')
plt.title("Películas vs Series en Netflix")
plt.show()
```
